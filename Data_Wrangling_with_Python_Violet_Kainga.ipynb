### Next Steps: Data Cleaning Steps

**Error in values**

- Extract the first integer value (lowest delay time) in the column "How_Long_Delayed"


**Missing Values**

- Impute the missing values in the field "How_Long_Delayed" with the mean value.


**Error in Datatypes**

- Convert "How_Long_Delayed" to int datatype.



**Error in field names**

- Rename the column "Boro" to "Borough".

```
# Lets first start by creating a copy of our dataframe 
# df_clean = df.copy(). We will use this copy as our cleaning copy.
# ---
#
BusBreakdown_df_clean = BusBreakdown_df.copy()
BusBreakdown_df_clean.head()
```

```
# Then extracting the lowest delay time in the column how_long_delayed from the string
# 
BusBreakdown_df_clean['How_Long_Delayed'] = BusBreakdown_df_clean['How_Long_Delayed'].str.extract('(\d+)')
BusBreakdown_df_clean['How_Long_Delayed'].head()
```

We impute the null values in 'how_long_delayed' column with mean of the column. This will take a couple of steps...

```
# We first convert our how_long_delayed to float type to allow for imputation
# ---
# 

BusBreakdown_df_clean['How_Long_Delayed']=BusBreakdown_df_clean['How_Long_Delayed'].apply(np.float64)
```

```
# Then later perform our mean imputation 
# ---
# 
BusBreakdown_df_clean['How_Long_Delayed']=BusBreakdown_df_clean['How_Long_Delayed'].fillna(BusBreakdown_df_clean['How_Long_Delayed'].mean())
```

```
# Then convert back our how_long_delayed column to integer datatype
# ---
# 
BusBreakdown_df_clean['How_Long_Delayed']=BusBreakdown_df_clean['How_Long_Delayed'].apply(np.int64)
```

```
# Then check for nulls in the column
# ---
# 
BusBreakdown_df_clean['How_Long_Delayed'].isnull().sum()
```

```
# Rename Boro column to Borough
# ---
# 
BusBreakdown_df_clean= BusBreakdown_df_clean.rename(columns= {'Boro' :'Borough'})
BusBreakdown_df_clean.columns
```

```
# Lastly we convert all values in our colums to lower case
# for ease of reading
# ---
# 
BusBreakdown_df_clean.columns=BusBreakdown_df_clean.columns.str.lower()
BusBreakdown_df_clean.columns
```


## 5. Solution Implementation

Here we investigate the questions that would help craft our recommendations.

### 5.a) Questions

```
# 1. Which bus companies that had the highest breakdowns?
# ---
#
breakdowns = BusBreakdown_df_clean.groupby(['bus_company_name']).count()

breakdowns.sort_values(by='busbreakdown_id', ascending=0)
```

```
# 2. What were the top 3 reasons for bus delays?
# ---
#
bus_delay = BusBreakdown_df_clean.groupby(['reason']).count()[['how_long_delayed']].reset_index()
bus_delay

# Sort to get the most frequent reason

bus_delay.sort_values(by='how_long_delayed', ascending=0)


```

```
# 3. How many students were in the buses when they broke down? 
# ---
BusBreakdown_df_clean.groupby(['number_of_students_on_the_bus']).count()[['busbreakdown_id']].reset_index()
```

```
# 4. Which were most frequent reasons for bus breakdowns?
# ---

breakdown_reasons = BusBreakdown_df_clean[BusBreakdown_df_clean.breakdown_or_running_late == 'Breakdown'].groupby(['reason']).count()


# Sort to get most frequent reasons
breakdown_reasons.sort_values(by='busbreakdown_id', ascending=0)


```

```
# 5. What were the most frequent reasons for the bus running late? 
# ---
Runninglate = BusBreakdown_df_clean[BusBreakdown_df_clean.breakdown_or_running_late == 'Running Late'].groupby(['reason']).count()

# Sort to get most frequent reasons
Runninglate.sort_values(by='busbreakdown_id', ascending=0)
```

```
# 6. What was the average delay time for each reason type?
# ---

avg_delay = BusBreakdown_df_clean.groupby('reason').mean().reset_index()

# Get the records with reasons and how long on average a delay took then sort 

reason_delay_avg=avg_delay[['reason', 'how_long_delayed']]

reason_delay_avg.sort_values(by='how_long_delayed', ascending=0)
```

### 5.b) Recommendations

From the above analysis, below are our recommendations:

1.   Late return from Field Trip is the major cause of delays
2.   Heavy traffic is the major cause of buses running late
3.  Mechanical problems causes the most bus breakdowns

4.G.V.C., LTD. Bus company had the most recorded breakdowns
